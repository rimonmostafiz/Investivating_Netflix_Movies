```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
netflix_df
```

```python
# Filter the data from netflix_df where type=Movie and release_year=1990s (1990-1999)
decade_1990 = np.logical_and(netflix_df["release_year"] >= 1990, netflix_df["release_year"] < 2000)
netflix_1990s_df = netflix_df[decade_1990]

# Filter by type Movie
netflix_1990s_movie_df = netflix_1990s_df[netflix_1990s_df["type"] == "Movie"]

# Show netflix_1990s_movie_df
netflix_1990s_movie_df
```

```python
# Filter all the durations
duration = netflix_1990s_movie_df["duration"].mode()[0]

# Generate histogram from duration
plt.hist(np.array(netflix_1990s_movie_df['duration']))

# Show histogram
plt.show()

# Show duration
print(duration)
```

```python
# Filter movies by genre and duration using numpy logical_and
netflix_short_action_df = netflix_1990s_movie_df[np.logical_and(
     netflix_1990s_movie_df["genre"] == "Action", 
     netflix_1990s_movie_df["duration"] < 90)]

# Show netflix_short_action_df
netflix_short_action_df
```
